```python
mean = df['Ex-Showroom_Price'].mean()

# Plot the bar graph for the mean price
plt.figure(figsize=(2,8))
plt.bar('Mean Ex-Showroom Price', mean, color='skyblue', edgecolor='black')

# Set the title and labels
#plt.title('Mean Ex-Showroom Price')
plt.ylabel('mean values in Ex-Showroom_Price')
plt.grid(axis='y')

plt.show()
```

```python
# frequency Method
# Define the threshold
Threshold = 0.05

# Calculate the normalized value counts for 'Ex-Showroom_Price'
Ex_Showroom_Price = df['Ex-Showroom_Price'].value_counts(normalize=True)

# Identify outliers based on the threshold
outliers = Ex_Showroom_Price[Ex_Showroom_Price < Threshold]

# Print the outliers
print('Outliers in the Model:')
print(outliers)
```
